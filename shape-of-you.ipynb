# Shape of you(r buildings)

Is there a difference between expert mappers interpretation of building footprints based on aerial imagery?

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.tools.plotting import scatter_matrix
from scipy.stats import gaussian_kde
from IPython.display import Image
from IPython.core.display import HTML
```

The data is a set of building polygons traced from aerial imagery (MS Bing).  The same location was traced by several for 15 minutes using JOSM.

**Description**

* `name` - name of mapper
* `sample` - building label for a polygon of the same building
* `area`= area in square meters
* `perim` - perimeter in meters
* `vertices` - number of vertices
* `shape`- "Circles will approach a value of 1 for a perfect circle.  Squares are around 0.78.  A thin thread-like object would have the lowest shape factor approaching 0." See [Morphometric & Densitometric Parameters](http://www.empix.com/NE%20HELP/functions/glossary/morphometric_param.htm)

```python
# Load the data, if you've forked my code, note that I'm using aboslute path here.  
# Replace the path to the location of your csv

data=pd.read_csv("/vagrant/data/shape_quality/building-team.csv")
data.head()
```

```python
# For each of the buildings, we get only those that were traced by at least 10 mappers.

sample_counts = data['sample'].value_counts().rename('sample_counts')
sampled = data.merge(sample_counts.to_frame(),
                                left_on='sample',
                                right_index=True)
sampled_10=sampled[sampled.sample_counts > 10]
```

```python
sampled_10.describe()
```

```python
# Get correlation matrix of values
sampled_10[['name','area','perim','shape','vertices']].corr()
```

```python
# Plot scatter matrix.
sns.pairplot(sampled_10[['name','area','perim','shape','vertices']], diag_kind='kde', kind='reg')
plt.show()
```

Now we see some basic trends here:
* Increase of `area` is proportional to its `perimeter`.
* `area` and `perim` are inversely correlated to the `shape` of a building.  The assumption here is that large buildings are more complex thus low `shape` while smaller building are more compact and thus simpler shape. 
* `shape` complexity significantly influence `vertices` count.

*We will focus now at the `shape` metrics.*  

```python
# Now, lets look at each mapper's metrics
sampled_10[['name','area','perim','shape','vertices']].groupby('name').describe()
```

```python
sampled_10.boxplot(by='name', column=['shape'], figsize=[15,8])
plt.show()
```

Hmmm, each mapper seems to trace building in varying degree of `shape` complexity.  What about the difference between the same buildings?

```python
# Boxplot for each sample building

sampled_10.boxplot(by='sample', column=['shape'], figsize=[15,8])
plt.show()
```

This is interesting, it seems that even with tracing the same building, each mapper interprest the shape very differently!
